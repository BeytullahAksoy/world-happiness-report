```python
year_2015 = pd.read_csv("2015.csv")
year_2016 = pd.read_csv("2016.csv")
year_2017 = pd.read_csv("2017.csv")
year_2018 = pd.read_csv("2018.csv")
year_2019 = pd.read_csv("2019.csv")
```

```python
happinnes_arr = []
happinnes_arr_first = []
happinnes_arr_last = []
```

```python
happinnes_arr.append(sum(year_2015["Happiness Score"]))
happinnes_arr.append(sum(year_2016["Happiness Score"]))
happinnes_arr.append(sum(year_2017["Happiness.Score"]))
happinnes_arr.append(sum(year_2018["Score"]))
happinnes_arr.append(sum(year_2019["Score"]))
```

```python
happinnes_arr_first.append(sum(year_2015[0:51]["Happiness Score"]))
happinnes_arr_first.append(sum(year_2016[0:51]["Happiness Score"]))
happinnes_arr_first.append(sum(year_2017[0:51]["Happiness.Score"]))
happinnes_arr_first.append(sum(year_2018[0:51]["Score"]))
happinnes_arr_first.append(sum(year_2019[0:51]["Score"]))
```

```python
happinnes_arr_last.append(sum(year_2015[-50:]["Happiness Score"]))
happinnes_arr_last.append(sum(year_2016[-50:]["Happiness Score"]))
happinnes_arr_last.append(sum(year_2017[-50:]["Happiness.Score"]))
happinnes_arr_last.append(sum(year_2018[-50:]["Score"]))
happinnes_arr_last.append(sum(year_2019[-50:]["Score"]))
```

```python
plt.plot([x for x in range(2015,2020)],happinnes_arr)
plt.title("Total Happiness Score of All Countries By Year")
plt.xlabel("Year")
plt.ylabel("Total Happiness Score")
plt.show()
```

```python
plt.plot([x for x in range(2015,2020)],happinnes_arr_first)
plt.title("Total Happiness Score of First 50 Countries")
plt.xlabel("Year")
plt.ylabel("Total Happiness Score")
plt.show()
```

```python
plt.plot([x for x in range(2015,2020)],happinnes_arr_last)
plt.xlabel("Year")
plt.ylabel("Total Happiness Score")
plt.title("Total Happiness Score of Last 50 Countries")
plt.show()
```

```python


a = year_2015[['Region','Happiness Score']].groupby('Region',as_index=False).mean()
a = a.sort_values('Happiness Score', ascending = False).reset_index(drop=True)
average = a["Happiness Score"].mean()
values = a["Happiness Score"]
values = values 
asOfDates = a['Region']

Value = a['Happiness Score'] / 100

fig, ax = plt.subplots()
ax.set_title('Happiness By Region')
ax.set_xlabel('Region')
ax.set_ylabel('Happiness Score')

plt.bar(asOfDates, values, color=['r' if v < average else 'g' for v in values])
ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

legend_handles = [Line2D([0], [0], linewidth=0, marker='o', markerfacecolor=color, markersize=12, markeredgecolor='none')
                  for color in ['g', 'r']]
ax.legend(legend_handles, ['More than average', 'Less than average'])
plt.xticks(rotation=90)
plt.show()
```
